```python
def custom_fn(x1, *l1):
    if x1=='AI':
        return np.sum(l1)
    elif x1=='Ai':
        return np.product(l1)
    elif x1=='cnc':
        return np.cumsum(l1)
    else:
        return 0
```

```python
df[['name','c1','c2','c2','c3','c4','c5','c6','c7']].apply(lambda x:custom_fn(*x),axis=1)
```

```python
df['checkvalidation']=df[['name','c1','c2','c2','c3','c4','c5','c6','c7']].apply(lambda x:custom_fn(*x),axis=1)
```

```python
def custom_fn(x1,x2,x3,x4,x5,x6):
    if x1=='AI':
        return np.sum(x1+x2+x3+x4+x5+x6)
    elif x1=='Ai':
        return np.product(x1+x2+x3+x4+x5+x6)
    elif x1=='cnc':
        return np.cumsum(x1+x2+x3+x4+x5+x6)
    else:
        return 0
```

```python
df['checkvalidation']=df[['name','c1','c2','c2','c3','c4','c5'].apply(lambda x:custom_fn(*x),axis=1)]
```
